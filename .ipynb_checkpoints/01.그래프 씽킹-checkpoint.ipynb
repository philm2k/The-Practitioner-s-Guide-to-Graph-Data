## 01. 그래프 씽킹

### 역사적 관점의 데이터베이스 기술변화 개요
> 1960 ~ 1980년대: 계층 데이터  
> 1980 ~ 2000년대: 개체-관계  
> 2000 ~ 2020년대: NoSQL  
> 2020년대 이후: 그래프

### 1.1 떠오르는 그래프 기술
#### 1.1.1 1960~1980년대: 계층 데이터
- 트리 형태의 자료구조로 데이터를 정리하는 것이 목표
- DBMS에서 레코드를 추출할 때 목표 세가지
> 기본키 사용  
> 순차적으로 모든 레코드 스캔  
> 한 레코드에서 다른 레코드로 탐색할 수 있는 연결 링크


### 1.1.2 1980~2000년대: 개체-관계
- 데이터 구조와 추출시스템을 분리 - Edgar F. Codd
- 데이터를 집합으로 조직화
##### 1990년대 후반에는 웹의 대중화를 통해 정보화 시대가 성큼 다가오면서 기존에 계획하거나 사용하지 않았던 형태의 데이터가 대거 등장했다.
- <u>배경 (인터넷의 대중화)</u>: 1990년대 후반 PC 통신에서 월드 와이드 웹(WWW)으로 전환되면서, 기업 중심의 데이터 생산 주체가 일반 대중으로 확장되었습니다.  
- <u>과거의 데이터</u>: 주로 은행 창구나 인사 시스템에서 쓰던 영수증, 사번, 금액 같은 정형 데이터(Structured Data)였습니다. 깔끔한 표(Table) 형태에 딱 들어맞는 데이터였죠.  
- <u>새로운 데이터</u>: 웹사이트의 로그(Log) 기록, 이메일 텍스트, 클릭 스트림(사용자가 웹페이지에서 클릭한 경로), 초창기 디지털 이미지와 오디오 등이 쏟아지기 시작했습니다. 이들은 정해진 규격이 없는 비정형/반정형 데이터였습니다.  

##### 이 시기에는 다양한 형태의 어마어마한 데이터가 애플리케이션에 쏟아졌다. 따라서 데이터의 용도를 사전에 정의할 수 없는 기존의 관계형 모델로는 모든 문제를 해결할 수 없음을 인지하기 시작했다. 
- <u>관계형 모델(RDBMS)의 한계</u>: 오라클(Oracle), MySQL 같은 관계형 데이터베이스는 데이터를 넣기 전에 반드시 스키마(Schema, 데이터 구조와 테이블 형태)를 완벽하게 설계해야 합니다.
- <u>'용도 사전 정의'의 어려움</u>: 새로 쏟아지는 웹 로그 데이터는 당장 어디에 어떻게 쓸지(마케팅 분석용인지, 시스템 오류 감지용인지) 예측하기 어려웠습니다. 용도를 모르니 미리 표(Table)의 칸을 짜둘 수가 없었던 것입니다.
- <u>비유</u>: 관계형 모델은 '칸막이가 딱 짜여진 서류함'과 같습니다. 규격 봉투(정형 데이터)는 착착 들어가지만, 모양이 제각각인 물건(웹 데이터)이 들어오면 서류함 자체를 새로 짜야 하거나 아예 넣지 못하는 구조적 한계가 있었습니다.

##### 업계에 상세한 저장 모델은 있었지만 데이터를 분석하거나 지능적으로 적용할 수는 없었다.
- <u>저장(Storage)과 분석(Analysis)의 괴리</u>: 당시 기술은 데이터를 유실 없이 '안전하게 기록(트랜잭션 처리)'하는 데 최적화되어 있었습니다. 데이터의 무결성(정확성)을 지키는 저장 모델은 훌륭했으나, 그 저장된 수천만 건의 데이터에서 의미 있는 패턴을 뽑아내는 '분석' 기능은 턱없이 부족했습니다.  
- <u>지능적 적용의 한계</u>: 당시는 지금의 머신러닝이나 고도화된 추천 알고리즘의 분석 환경(하두업, 스파크 등)이 대중화되기 전이었습니다. 데이터가 그냥 대용량 창고에 쌓여만 있는 '죽은 데이터'였을 뿐, 이를 기반으로 소비자의 행동을 예측하거나 지능적인 비즈니스 의사결정에 활용(비즈니스 인텔리전스, BI)하기에는 연산 속도와 분석 모델 측면에서 기술적 정체기가 있었습니다.
