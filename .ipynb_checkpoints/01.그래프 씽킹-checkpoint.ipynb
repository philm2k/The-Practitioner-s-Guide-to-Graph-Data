## 01. 그래프 씽킹

### 역사적 관점의 데이터베이스 기술변화 개요
> 1960 ~ 1980년대: 계층 데이터  
> 1980 ~ 2000년대: 개체-관계  
> 2000 ~ 2020년대: NoSQL  
> 2020년대 이후: 그래프

### 1.1 떠오르는 그래프 기술
#### 1.1.1 1960~1980년대: 계층 데이터
##### 등장 배경
> IBM의 IMS(Information Management System)로 대표되는 계층형 DB는 컴퓨터가 기업 환경에 막 도입되던 시기에 탄생했다. 당시의 데이터는 군수물자 관리, 급여 대장처럼 자연스럽게 트리(tree) 구조를 띠는 경우가 많았고, 이를 그대로 모델링한 것이 계층형 DB다.
##### 구조적 특징
> - 데이터는 부모-자식(parent-child) 관계로만 연결  
> - 각 자식 노드는 반드시 하나의 부모만 가짐
> - 데이터 탐색은 루트(root)에서 시작해 아래로만 이동
> - 트리 형태의 자료구조로 데이터를 정리하는 것이 목표
##### 극복한 결점
> 당시 파일 시스템(flat file)에 비해 비약적인 발전이었다. 
> 데이터를 개별 파일에 분산 저장할 때와 달리, 관련 데이터를 계층 구조로 묶어 중복을 줄이고 접근 경로를 명확히 했다.
##### 남겨진 결점 — 다음 세대가 극복해야 할 문제
> - 다대다(M:N) 관계 불가: "직원이 여러 프로젝트에 속한다"를 표현할 수 없음. 데이터를 중복 저장하거나 포기해야 함
> - 경직된 스키마: 부모를 변경하거나 구조를 수정하면 전체 트리를 재구성해야 함
> - 탐색 경로 고정: 반드시 루트에서부터 순차 탐색. 원하는 리프 노드에 바로 접근 불가
> - 데이터 종속성: 물리적 저장 구조와 논리적 구조가 결합되어 있어, 저장 방식이 바뀌면 애플리케이션도 수정해야 함

### 1.1.2 1980~2000년대: 개체-관계
#### 등장 배경과 핵심 혁신
> 에드가 코드(Edgar F. Codd)가 1970년 IBM 논문에서 **관계 모델(relational model)**을 제안한 것이 출발점이다.  
> 핵심 아이디어는 단순하지만 혁명적이었다:  
> 데이터를 행(row)과 열(column)로 이루어진 테이블에 저장하고, 물리적 저장 방식과 논리적 구조를 완전히 분리하자.

##### 계층형 DB의 결점을 극복한 방식
- 다대다 관계 해결 — 조인(JOIN)과 외래키(Foreign Key)
> 계층형에서 불가능했던 복잡한 관계를 중간 테이블(junction table)과 조인으로 자연스럽게 표현한다.
- 데이터 독립성 확보 — SQL의 선언적 질의
> "어떻게 찾을지(HOW)"가 아닌 "무엇을 찾을지(WHAT)"만 기술하면, 실행 계획은 DB 엔진이 결정한다. 물리적 저장 방식이 바뀌어도 SQL은 그대로다.
- 정규화(Normalization)로 데이터 무결성 확보
> 1NF~3NF의 정규화 규칙으로 중복을 제거하고, ACID 트랜잭션(원자성·일관성·격리성·지속성)으로 데이터 정합성을 보장한다. 은행 이체처럼 "출금과 입금이 반드시 함께 성공하거나 실패해야 하는" 연산을 안전하게 처리할 수 있게 됐다.

##### 남겨진 결점 — 다음 세대가 극복해야 할 문제
- 수직 확장의 한계
> 트래픽 증가 시 서버 스펙을 높이는 것(Scale-Up)만 가능. 서버를 수평으로 늘리는(Scale-Out) 것이 구조적으로 어려움
- 고정 스키마
> 서비스 중에 컬럼을 추가·변경하면 전체 테이블에 락(lock)이 걸려 대규모 시스템에서 치명적
- 임피던스 불일치
> 객체지향 언어의 클래스 계층 구조와 테이블의 평면 구조가 달라 ORM 같은 매핑 레이어가 필요
- 비정형 데이터 취약
> JSON, 이미지, 로그처럼 구조가 일정하지 않은 데이터를 테이블에 억지로 끼워 맞춰야 함
- CAP 정리의 제약
> 분산 환경에서 일관성(Consistency)을 지키기 위해 가용성(Availability)이나 분할 내성(Partition Tolerance)을 희생

##### 1990년대 후반에는 웹의 대중화를 통해 정보화 시대가 성큼 다가오면서 기존에 계획하거나 사용하지 않았던 형태의 데이터가 대거 등장했다.
- 배경 (인터넷의 대중화)
> 1990년대 후반 PC 통신에서 월드 와이드 웹(WWW)으로 전환되면서, 기업 중심의 데이터 생산 주체가 일반 대중으로 확장되었습니다.
- 과거의 데이터
> 주로 은행 창구나 인사 시스템에서 쓰던 영수증, 사번, 금액 같은 정형 데이터(Structured Data)였습니다. 깔끔한 표(Table) 형태에 딱 들어맞는 데이터였죠.  
- 새로운 데이터
> 웹사이트의 로그(Log) 기록, 이메일 텍스트, 클릭 스트림(사용자가 웹페이지에서 클릭한 경로), 초창기 디지털 이미지와 오디오 등이 쏟아지기 시작했습니다. 이들은 정해진 규격이 없는 비정형/반정형 데이터였습니다.  

##### 업계에 상세한 저장 모델은 있었지만 데이터를 분석하거나 지능적으로 적용할 수는 없었다.
- <u>저장(Storage)과 분석(Analysis)의 괴리</u>: 당시 기술은 데이터를 유실 없이 '안전하게 기록(트랜잭션 처리)'하는 데 최적화되어 있었습니다. 데이터의 무결성(정확성)을 지키는 저장 모델은 훌륭했으나, 그 저장된 수천만 건의 데이터에서 의미 있는 패턴을 뽑아내는 '분석' 기능은 턱없이 부족했습니다.  
- <u>지능적 적용의 한계</u>: 당시는 지금의 머신러닝이나 고도화된 추천 알고리즘의 분석 환경(하두업, 스파크 등)이 대중화되기 전이었습니다. 데이터가 그냥 대용량 창고에 쌓여만 있는 '죽은 데이터'였을 뿐, 이를 기반으로 소비자의 행동을 예측하거나 지능적인 비즈니스 의사결정에 활용(비즈니스 인텔리전스, BI)하기에는 연산 속도와 분석 모델 측면에서 기술적 정체기가 있었습니다.

### 1.1.3 2000~2200년대: NoSQL
#### 등장 배경
> 구글, 아마존, 페이스북 같은 인터넷 기업들이 수십억 사용자를 다루면서 관계형 DB의 수직 확장 한계에 부딪혔다. "완벽한 일관성보다는 언제나 응답 가능한 시스템"이 필요해졌고, 동시에 SNS 게시물·로그·센서 데이터 같은 비정형 데이터가 폭발적으로 늘어났다. NoSQL은 단일 기술이 아니라, 이 문제를 각각 다른 방식으로 해결하는 여러 데이터 모델의 집합이다.

#### RDBMS의 결점을 극복한 방식
- 수평 확장(Scale-Out) — 분산 아키텍처
> 데이터를 여러 노드에 분산·복제하는 **샤딩(Sharding)**과 **복제(Replication)**를 기본 설계에 내장했다. 트래픽이 늘면 서버를 추가하면 그만이다.
- 유연한 스키마 — 스키마리스(Schema-less)
> MongoDB 같은 문서형 DB는 JSON 유사 구조를 그대로 저장한다. 컬럼이 없어도 되고, 문서마다 필드가 달라도 된다.
- 데이터 모델 다양화 — 용도에 맞는 선택
> - 키-값(Key-Value): Redis, 초고속 읽기/쓰기세션 캐시, 실시간 순위
> - 문서형(Document): MongoDB, 유연한 스키마, 중첩 구조콘텐츠, 사용자 프로파일
> - 컬럼형(Column-family): Cassandra, 대용량 시계열 데이터IoT, 로그, 금융 거래 이력
> - 그래프(Graph): Neo4j, 관계 탐색소셜 네트워크, 추천
- BASE 철학 — 가용성 우선
> ACID 대신 BASE(Basically Available, Soft state, Eventually consistent)를 채택. 순간적 불일치를 허용하되, 시간이 지나면 일관성에 수렴한다. 쇼핑몰 재고 표시처럼 "약간의 오차는 허용하되 항상 응답해야 하는" 서비스에 적합하다.

#### 남겨진 결점
- 복잡한 조인 불가
> 관계형 조인이 없어, 연관 데이터를 합치는 로직을 애플리케이션 레이어에서 처리해야 함
- 트랜잭션 보장 약함
> 최종 일관성(Eventual Consistency)으로는 금융·의료 등 강한 정합성이 필요한 도메인에 부적합
- 표준 질의 언어 부재
> DB마다 API가 달라 학습 비용이 높고 이식성이 낮음

### 1.1.4 2020년대 이후 : 그래프

#### 극복한 결점: 연결 데이터의 관계 탐색
그래프 DB는 NoSQL의 한 유형으로 분류되기도 하지만, 관계 탐색 문제를 해결하는 독자적 패러다임으로 4세대로 구분하는 것이 적절하다. Neo4j, TigerGraph, Amazon Neptune, DataStax가 대표적이다.
- 관계를 1등 시민(First-Class Citizen)으로
> 관계형 DB에서 관계는 외래키와 JOIN이라는 간접적 수단으로 표현된다. 그래프 DB에서 엣지(Edge, 관계)는 노드(Node, 엔티티)와 동등한 지위를 갖는 독립적 데이터 객체다. 엣지에 방향·가중치·속성을 직접 부여할 수 있다.
- Index-Free Adjacency
> 각 노드는 인접 노드들의 포인터를 직접 보유한다. 관계를 찾을 때 인덱스를 검색하거나 JOIN 연산을 수행할 필요 없이, 포인터를 따라가면 O(1)에 인접 노드에 도달한다. 그래프의 크기(전체 노드·엣지 수)와 무관하게 탐색 성능이 일정하게 유지된다.
- 다중 홉 탐색의 선형 성능
> "친구의 친구의 친구까지 3단계 연결"을 관계형 DB로 처리하면 세 번의 JOIN이 필요하고 데이터 규모에 따라 성능이 지수적으로 저하된다. 그래프 DB는 엣지를 따라 포인터를 순회하므로 홉 수에 비례하는 선형 시간에 처리한다.
- 직관적 모델링
> 현실 세계의 네트워크 구조(소셜 관계, 공급망, 지식 체계)를 노드와 엣지로 그대로 표현한다. 객체지향 모델과 저장 모델 사이의 임피던스 불일치(Impedance Mismatch)가 최소화된다.

#### 구조적 특징
> 데이터는 **노드(Node)·엣지(Edge)·속성(Property)**의 세 요소로 구성된다. 노드는 사람·장소·개념 같은 엔티티를 표현하고, 엣지는 노드 간 유향 관계를 나타내며, 속성은 노드와 엣지에 부착된 키-값 데이터다. 질의는 Cypher(Neo4j)나 Gremlin(Apache TinkerPop)으로 수행한다.

#### 남긴 결점: 그래프 DB의 한계
그래프 DB 역시 트레이드오프에서 자유롭지 않다.
- 수평 확장의 구조적 어려움
> 노드 간 엣지가 파티션을 가로지르는 경우 탐색 성능이 급격히 저하된다.
> 대규모 분산 그래프 처리는 관계형이나 NoSQL 대비 여전히 어렵다.
- 표준 질의 언어의 부재
> Cypher, Gremlin, SPARQL 등 DB마다 다른 질의 언어를 사용해 이식성이 낮다.
> ISO GQL 표준 제정이 진행 중이나 아직 보편화 전이다.
- 집계·분석 연산의 취약성
> SUM, GROUP BY 같은 테이블 전체 스캔 집계는 관계형 DB 대비 비효율적이다.
> OLAP 성격의 대용량 분석에는 부적합하다.
- 높은 학습 곡선
> 그래프 사고(Graph Thinking)로의 패러다임 전환, Cypher/Gremlin 습득, 노드-엣지 모델링은 기존 SQL 개발자에게 상당한 진입 장벽이다.
- 생태계 미성숙
> MySQL·PostgreSQL 대비 커뮤니티·클라우드 지원·BI 연동 도구가 아직 제한적이다.
